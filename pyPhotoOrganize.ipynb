## Photo Organizer

- point to photo top (phDIR) directory : contains all photos
- create an inventory of ALL photos within phDIR
- attrib per photo (<phDIR>/<dir>/<bn>)
    - bn : basename  (fn.ext)
    - dt : date of photo (image, or mtime of file)
    - ext : extension
    - dir : relative path from phDIR to bn
